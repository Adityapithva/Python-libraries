<h1>Pyplot API</h1>

```python
data = {
    'Salary':[20000,30000,40000,50000,65000,72000,80000,85000,67000,75000,90000,95000]
}
df = pd.DataFrame(data)
df
```

```python
df.shape
```

```python
# Line Plot
plt.plot(df['Salary'],color='red',marker='o',linestyle='--')
plt.grid()
plt.show()
```

```python
#Histogram
plt.hist(df['Salary'],bins=5,color='green')
plt.show()
```

```python
# Box plot
plt.boxplot(df['Salary'])
plt.plot()
```
